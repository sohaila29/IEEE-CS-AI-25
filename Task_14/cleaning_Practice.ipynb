```python
#Correct syntax errors in columns
df_users['Name'] = df_users['Name'].str.replace(r'[^a-zA-Z\s]', '', regex=True)
```

```python
df_users = df_users.rename(columns={"Name": "First_name"})
```

```python
#Set a unique pattern for a column
def format_phone(phone):
    if pd.isna(phone) or str(phone).lower() == 'nan':
        return "None"
    digits = ''.join(filter(str.isdigit, str(phone)))
    
    if len(digits) == 10:
        return f"{digits[:3]}-{digits[3:6]}-{digits[6:]}"
    
    return "None"

df_users['Phone'] = df_users['Phone'].apply(format_phone)
df_users
```

```python
#Delete rows with None values
df_users = df_users.replace('None', pd.NA)
df_users = df_users.dropna()
df_users
```
